```python
def diff_img(a, b):
    a = a.reshape(-1).astype(np.float)
    b = b.reshape(-1).astype(np.float)
    diff = a-b
    sq_diff = diff**2
    sq_diff_sum = np.mean(sq_diff)
    return np.sqrt(sq_diff_sum)
```

```python
diff_img(new_tmp_1, old_tmp_1)
```
